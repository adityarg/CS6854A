```python
def edit_distance(str1, str2):
    m=len(str1); n=len(str2)
    dp = [[0 for x in range(n + 1)] for x in range(m + 1)]
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0:
                dp[i][j] = j
            elif j == 0:
                dp[i][j] = i
            elif str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i][j-1],dp[i-1][j],dp[i-1][j-1])
    return dp[m][n]

# ed=2
# for i in dist:
#     for j in ddata:
#         if len(i)>=len(j) and edit_distance(j,i)<=ed and i!=j and i not in ddata and j not in dist:
#             print(i,"\t",j)
#         elif len(i)<=len(j) and edit_distance(i,j)<=ed and i!=j and i not in ddata and j not in dist:
#             print(i,"\t",j)
for i in ddata:
    if i not in dist:
        print(i)
```
